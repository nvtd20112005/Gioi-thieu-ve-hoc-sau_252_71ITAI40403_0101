```python
'''BT1: cho y = 5x^5 + 6x^3 - 3x + 1 Cho biết độ dốc của đa thức trên ở điểm nào '''
import torch

x = torch.tensor(2.0, requires_grad=True)
y = 5*x**5 + 6*x**3 - 3*x + 1
y.backward()

print(x.grad)
```

```python
'''
BT2: tạo 1 tensor ban đầu có giá trị là 2 Định nghĩa hàm số và 
tính gradient Y = x^3 + 2x^2 + 5x + 1 Hãy tính dy/dx tại giá trị của x 
Dùng phương pháp gradient descent với learning rate alpha = 0.1 để cập nhật giá trị x trong 50 vòng lặp 
'''
# Y = x^3 + 2x^2 + 5x + 1
# 1 tensor ban đầu có giá trị là 2 (x = 2)
# learingrate alpha = 0.1   
import torch

x = torch.tensor(2.0, requires_grad=True)
alpha = 0.1

for _ in range(50):
    if x.grad is not None:
        x.grad.zero_()
        
    y = x**3 + 2*x**2 + 5*x + 1
    y.backward()
    
    with torch.no_grad():
        x -= alpha * x.grad

print("x =", x.item())
print("dy/dx =", x.grad.item())

```

```python
'''
BT3: tạo 1 tập dữ liệu giả lập với x là số giờ học( ngẫu nhiên từ 1 - 10) và y là số điểm đ… theo công thức y =3x + 5 + noise 
Với noise là 1 giá trị ngẫu nhiên nhỏ 
1 khởi tạo tham số ư và b ngẫu nhiên với requires_grad = True 
2 tính MSE 3 tính gradient 
4 cập nhật tham số w và b bằng gradient descent với learning rate alpha = 0.1
5 lặp lại quá trình trên trong 100 vòng lặp để xem sự hội tụ của nó '''
import torch

# tạo dữ liệu giả lập
x = torch.randint(1, 11, (100, 1)).float()      # số giờ học
noise = torch.randn(100, 1) * 0.5               # noise nhỏ
y = 3*x + 5 + noise                             # số điểm

# khởi tạo tham số w, b
w = torch.randn(1, requires_grad=True)
b = torch.randn(1, requires_grad=True)

alpha = 0.1

# gradient descent 100 vòng lặp
for _ in range(100):
    # dự đoán
    y_pred = w*x + b
    
    # tính MSE
    loss = ((y_pred - y)**2).mean()
    
    # tính gradient
    loss.backward()
    
    # cập nhật w, b
    with torch.no_grad():
        w -= alpha * w.grad
        b -= alpha * b.grad
    
    # reset gradient
    w.grad.zero_()
    b.grad.zero_()

print("w =", w.item())
print("b =", b.item())

```

Sở dĩ ở trường hợp trên tensor có giá trị 99 còn ở trường hợp dưới thì không, là do cách PyTorch tạo tensor khác nhau. 

Khi dùng `torch.from_numpy(arr)`, tensor và mảng NumPy dùng chung vùng nhớ, nên khi giá trị của `arr` thay đổi thì tensor cũng thay đổi theo. 

Ngược lại, khi dùng `torch.tensor(arr)`, PyTorch tạo ra một tensor mới bằng cách sao chép dữ liệu, nên sau khi tạo tensor, việc thay đổi mảng NumPy không ảnh hưởng đến tensor, vì hai bên không còn liên kết bộ nhớ với nhau.


```python
'''
BT5 : bạn giúp thầy tạo tensor với 
-empty 
-zeros 
-ones 
-random 
-reshape với view và view as
'''
import torch

a = torch.empty(2, 3)
b = torch.zeros(2, 3)
c = torch.ones(2, 3)
d = torch.rand(2, 3)

x = torch.arange(12)
y = x.view(3, 4)
z = x.view_as(y)

print(a)
print(b)
print(c)
print(d)
print(y)
print(z)
```
